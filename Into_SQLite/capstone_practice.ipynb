```python
import pandas as pd
import sqlite3
dbfile = "Combination_lab_elections.db"

con08 = sqlite3.connect(dbfile)
DF_2008 = pd.read_sql_query('select * from Transfer_Table_2008',con08)

DF_2008[["Apportionment_pop", "State_Vote_Total"]] = DF_2008[["Apportionment_pop", "State_Vote_Total"]].apply(pd.to_numeric)

DF_2008

Alaske_count = 0

for A_row in DF_2008.itertuples():
    if A_row.STATE == "Alaska":
        Alaske_count += 1

print(Alaske_count)
DF_2008 = DF_2008.assign(Remander=DF_2008["Apportionment_pop"] / DF_2008["Num_Reps_Capped"])
DF_2008.Remander = DF_2008.Remander.round()
DF_2008
```

## EC NE style 
 - state vote total (FEC election)/number of reps (apportment table by year) = pop_per_vote
 - apportion by candidates votes for state (FEC table) dividing by pop_per_vote
 - use remander divide for candidate remanders and given vote to most of remanders if remander pop > pop_per_vote
 - give 2 additional votes to top number of electorate votes for senate seats

## EC full divide
 - state vote total (FEC election)/(number of reps (apportment table by year) + 2 for senators) = pop_per_vote
 - apportion by candidates votes for state (FEC table) dividing by pop_per_vote
 - use remander divide for candidate remanders and given vote to most of remanders if remander pop > pop_per_vote
